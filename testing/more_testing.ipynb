```python
word_list_train = "data/250k_train.txt"
word_list_val = "data/250k_test.txt"
number_games = 10
model = Model("big_embed")
```

```python
model.to('cpu')
# training_data_generator = Train_on_Batch(word_list_train, model, 10000)
test_data_generator = Train_on_Batch(word_list_val, model, 500) 
validate_set = test_data_generator.generate_data()
```

```python
train_for = 5000 #! games
model.to('cpu')
for i in range(500, train_for+500, 500):
    #! generate data for 500 games at a time, train on it and mve ahead.
    
    t_d_g = Train_on_Batch(word_list_train, model, i, verbose=False)
    training_data = t_d_g.generate_data()
    model.to('cuda')
    training_data = [i.to('cuda') for i in training_data]
    train_one_batch(model, training_data[0], training_data[1], training_data[2], 1000, torch.optim.NAdam(model.parameters(), lr = 5e-3))
    model.to('cpu')
    # del model
    gc.collect()
    torch.cuda.empty_cache()

    
    # pbar.set_description("%.3f Val Loss" % )
    
```
